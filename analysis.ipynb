```python
import pandas as pd
```

```python
file_path = 'Badeel_Data.csv'
df = pd.read_csv(file_path, skiprows=1)
```

```python
print(df.head())
print(df.columns)
```

# Data preprocessing

```python
# Remove empty columns
df = df.dropna(axis=1, how='all')
```

```python
df.columns = ['Date', 'Name', 'English Name', 'Manufacturer', 'Details', 'Type', 'Reviewed', 'Alternate', 'Additional Info', 'Type Again']
```

```python
df.reset_index(inplace=True)
df.rename(columns={'index': 'Original Index'}, inplace=True)
```

```python
print("Cleaned DataFrame:")
display(df)
```

### Work with the duplicated data

```python
# Find duplicated name
duplicate_names = df[df.duplicated(['Name'], keep=False)]
print("Duplicate Names:")
display(duplicate_names)
```

```python
# Sort duplicate name by Name
print("Duplicate Names Sorting by Name:")
duplicate_names.sort_values(by='Name')
```

```python
# Export duplicates in a csv form
output_file_path = 'duplicates.csv'
duplicate_names.sort_values(by='Name').to_csv(output_file_path, index=False, encoding='utf-8-sig')

print(f"Duplicate rows have been exported to {output_file_path}")
```

```python
def merge_details(group):
    return pd.Series({
        'Original Index': ', '.join(map(str, group['Original Index'].tolist())),
        'Date': ', '.join(group['Date'].dropna().astype(str).unique()),
        'English Name': ', '.join(group['English Name'].dropna().astype(str).unique()),
        'Manufacturer': ', '.join(group['Manufacturer'].dropna().astype(str).unique()),
        'Details': ' | '.join(group['Details'].dropna().astype(str).unique()),
        'Type': ', '.join(group['Type'].dropna().astype(str).unique()),
        'Reviewed': ', '.join(group['Reviewed'].dropna().astype(str).unique()),
        'Alternate': ', '.join(group['Alternate'].dropna().astype(str).unique()),
        'Additional Info': ' | '.join(group['Additional Info'].dropna().astype(str).unique()),
        'Type Again': ', '.join(group['Type Again'].dropna().astype(str).unique()),
    })
```

```python
# merged_duplicates variable
merged_duplicates = duplicate_names.groupby('Name').apply(merge_details).reset_index()
```

```python
# Export duplicates by mergeing their detail as a csv form
output_file_path = 'merged_duplicates.csv' 
merged_duplicates.to_csv(output_file_path, index=False, encoding='utf-8-sig')

print(f"Merged duplicate rows have been exported to {output_file_path}")
```

```python
# Remove original duplicates from the DataFrame
non_duplicates = df[~df.duplicated(['Name'], keep=False)]

# Combine the non-duplicate rows with the merged duplicate rows
final_df = pd.concat([non_duplicates, merged_duplicates], ignore_index=True)
```

```python
# Sort by index
final_df['Original Index'] = final_df['Original Index'].apply(lambda x: int(x.split(',')[0]) if isinstance(x, str) else x)
final_df = final_df.sort_values(by='Original Index')
```

```python
# Export Final df without duplicate as a csv form
output_file_path = 'final_merged_data.csv'
final_df.to_csv(output_file_path, index=False, encoding='utf-8-sig')

print(f"Final df with merged duplicates has been exported to {output_file_path}")
```

### Filtering data by types

```python
def filter_by_type(df, type_value):
    return df[df['Type'] == type_value]
```

```python
# Filter by 'boycott' 
boycott_df = filter_by_type(df, 'boycott')
print("Filtered by 'boycott' Type:")
display(boycott_df)

# Export as a csv form
output_file_path = 'Type_boycott.csv' 
boycott_df.to_csv(output_file_path, index=False, encoding='utf-8-sig')
print(f"Feltering by Type 'boycott' have been exported to {output_file_path}")
```

```python
# Filter by 'alternative' 
alternative_df = filter_by_type(df, 'alternative')
print("Filtered by 'alternative' Type:")
display(alternative_df)

# Export as a csv form
output_file_path = 'Type_alternative.csv' 
alternative_df.to_csv(output_file_path, index=False, encoding='utf-8-sig')
print(f"Feltering by Type 'alternative' have been exported to {output_file_path}")
```
